```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Initialize Chrome webdriver
driver = webdriver.Chrome()
url = "http://www.sba.oakland.edu/faculty/isken/courses/mis6900_s24/getting_data_from_web.html"

# Get webpage content using Selenium
driver.get(url)

# Get page source after dynamic content loading
page_source = driver.page_source

# Close the Selenium webdriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Function to print all elements and their text data recursively
def print_elements_with_text(element, level=0):
    if element.name is not None:
        print("  " * level + element.name + ": " + element.get_text().strip())
    for child in element.children:
        if child.name is not None:
            print_elements_with_text(child, level + 1)

# Start printing elements with text from the root
print_elements_with_text(soup)
```
